# Krill popualtion data formating and exploration

1. The model KILLPODYM

The model KRILLPODYM  is presented in the article Green et al 2023: https://doi.org/10.3389/fmars.2023.1218003 
It is a mechanistic model of the krill population at the circumpolar scale and it is structured in life stages (Young larvae; free larvae; late larvae, juvenile and adult).
Here we used only the adult life stage as we assume they contribute the most to POC flux with the fecal pellets. 

2. Formating

This output from KRILLPODYM represents the krill biomass density [g/m²] over the Southern Ocean for all the life stages. We select the adults. The spatial resolution is not the same as the resolution of POC attenuation and sequetsration time as well as of the fishing data.
Therefore we can put them in the "harmonised resolution" of 2°x2° longitude-latitude grid, with center of the cell on odd numbers and the longitude in a Atlantic centered format (from -180 to 180).

```python
# open the .nc files with the model output of the four krill life stages
# this file corresponds to the last iteration of the spin up (results presented in the paper)
krill_biomass = xr.open_dataset('krill_biomass_20101229.nc')
krill_biomass
```

```python
krill_biomass.to_dataframe().describe()
```

```python
# we log the data to have a more readable visualization
def my_log(x):
    return np.round(np.log10(x+1), 5)

krill_biomass = krill_biomass.assign(krill_adult_log = my_log(krill_biomass['krill_adults']))
```

```python
map_proj = ccrs.SouthPolarStereo()

p = krill_biomass["krill_adult_log"].plot(
    transform=ccrs.PlateCarree(),  # the data's projection
    add_colorbar = False,
    subplot_kws={"projection": map_proj},
    cmap= "viridis",#"gnuplot2"
    #title = 'mean time of carbon sequesration'
)  # the plot's projection

#p.axes.coastlines()
p.axes.gridlines()
p.axes.add_feature(cfeature.LAND, color='grey')
p.axes.add_feature(cfeature.COASTLINE)
cb = plt.colorbar(p, label = 'log(krill_biomass_density + 1) [g/m²]', shrink = 0.8, boundaries = np.linspace(0,5,21))
plt.title('Map of krill density - KRILLPODYM')
```

## Change resolution

In order to have the same resolution with the other data (POC attenution/sequestration time) and fishing data we change the resolution from 0.25°x0.25° to 2°x2° grid like the other datasets we use in the spatail analysis (as OCIM data).

```python
# interpolate with the same lat long vectors that we used for envir data 
# create desired latitude and longitude vectors (start at -79 to match the envir and OCIM coords wich are 2°x2° on odd numbers)
new_lat = np.arange(-79, -39, 2) 
new_lon = np.arange(-179, 180, 2)

krill_biomass_interpolated = krill_biomass.interp(latitude=new_lat, longitude=new_lon)

# calcul again the log of krill density on the interpolated data (if not, it changes a lot!)
krill_biomass_interpolated = krill_biomass_interpolated.assign(krill_adult_log = my_log(krill_biomass_interpolated['krill_adults']))
```

```python
map_proj = ccrs.SouthPolarStereo()

p = krill_biomass_interpolated["krill_adult_log"].plot(
    transform=ccrs.PlateCarree(),  # the data's projection
    add_colorbar = False,
    subplot_kws={"projection": map_proj},
    cmap= "viridis"
) 

p.axes.gridlines()
p.axes.add_feature(cfeature.LAND, color='grey')
p.axes.add_feature(cfeature.COASTLINE)
cb = plt.colorbar(p, label = 'Krill biomass density [g/m²] (in scale log(x+1))', shrink = 0.8, boundaries = np.linspace(0,3.5,36))
plt.title('Map of krill density - KRILLPODYM')
```

```python
# transform also into dataframe
krill_biomass_df = krill_biomass_interpolated.to_dataframe()
krill_biomass_df.describe()
```
