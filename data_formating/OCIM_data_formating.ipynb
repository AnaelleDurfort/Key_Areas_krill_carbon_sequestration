# Formating of the time of sequestration

The time of sequestration is calculated by the Invert Circulation Ocean Model (OCIM) (Devries & Holzer (2019); Devries et al (2014)). It was retreive from the website https://tdevries.eri.ucsb.edu/models-and-data-products/.

This data is on a 2°x2° spatial grip with 48 depth layers. We will use this resolution for all the orther data and to run the overlapp analysis because it is the lowest resolution we have (temperature, oxygene, krill population and fishing are on finer resolution, so we will aggregate them to the resolution of these OCIM data).

However we have to change:
- the latitude because the OCIM uses a 91 degres of latitude instead of 90. So we have to interpolate the data on a new regular grid of 2° interval (instead of 1.98.

- the longitude because uses a "Pacific centered" longitude with values from 1 to 360 degres. Instead, we prefer the Atlentic centered longitude with values from -180 to 180 because all the other datasets use this format.

```python
# import libraries (ok sous base anaconda)
import matplotlib.pyplot as plt
import xarray as xr
import numpy as np
import pandas as pd
import os
import scipy
import math
```

```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
```

```python
# To be change with local directory of the data (file too heavy to be on git)
os.chdir('C:/Users/anael/Documents/PhD/4_krill_fisheries_Carbon_seq_time/Git_repo/Key_Areas_krill_carbon_sequestration/data_envir')
fseq_OCIM2_48L = xr.open_dataset('fseq_OCIM2_48L.nc')
fseq_OCIM2_48L # the lat, long and depth are in coordinates but in variables
```

```python
fseq_OCIM2_48L = fseq_OCIM2_48L.assign_coords(longitude = fseq_OCIM2_48L.LON.sel(latitude=0, depth=0),
                                              latitude = fseq_OCIM2_48L.LAT.sel(longitude=0, depth=0), 
                                             depth = fseq_OCIM2_48L.DEPTH.isel(longitude=0, latitude=0))
fseq_OCIM2_48L
```

```python
# plots are inverted => transpose the data 
fseq_OCIM2_48L = fseq_OCIM2_48L.transpose()
fseq_OCIM2_48L['mean_seq_time'].mean(dim = 'depth').plot()
```

```python
#interpolate with new lat 
# /!\ takes a few seconds
newlat = list(np.arange(-89, 90, 2))
fseq_OCIM2_48L_interpolated = fseq_OCIM2_48L.interp(latitude = newlat)
fseq_OCIM2_48L_interpolated
```

```python
# Interpolate with new longitude that is atlantic centered
long_pac = np.asarray(fseq_OCIM2_48L_interpolated.longitude)
long_atl = ((long_pac - 180) % 360) - 180
long_atl
```

```python
fseq_OCIM2_48L_interpolated = fseq_OCIM2_48L_interpolated.assign_coords(longitude = long_atl)
# Then we have to sort the xarrays by longitude, otherwise it cannot plot the data 
# /!\ Do not run these line multiple times => shift to the atlantic centered representation again
fseq_OCIM2_48L_interpolated  = fseq_OCIM2_48L_interpolated.loc[{'longitude':sorted(fseq_OCIM2_48L_interpolated.coords['longitude'].values)}]
```

```python
fseq_OCIM2_48L_interpolated # ok longitude from -179 to 179
```

```python
# Check with mean time seq
# compare maps of interpolated and original data to check the interpolation
time_seq_interpolated = fseq_OCIM2_48L_interpolated['mean_seq_time']
time_seq = fseq_OCIM2_48L['mean_seq_time']


time_seq_interpolated.mean(dim = 'depth').plot()
plt.show()
time_seq.mean(dim = 'depth').plot()
```

#### Crop the data to keep the Southern Ocean only

All the data southward than -40°S are selected.

```python
fseq_OCIM2_48L_SO= fseq_OCIM2_48L.sel(latitude=slice(-80, -40))
fseq_OCIM2_48L_interpolated_SO = fseq_OCIM2_48L_interpolated.sel(latitude=slice(-80, -40))
```

```python
# This is the map projection we want to plot *onto*
map_proj = ccrs.SouthPolarStereo()

p = fseq_OCIM2_48L_SO['mean_seq_time'].mean(dim = 'depth').plot(
    transform=ccrs.PlateCarree(),  # the data's projection
    add_colorbar = False,
    subplot_kws={"projection": map_proj},
    #cmap="PuBu",
    #title = 'mean time of carbon sequesration'
)  # the plot's projection

#p.axes.coastlines()
p.axes.gridlines()
p.axes.add_feature(cfeature.LAND, color='grey')
p.axes.add_feature(cfeature.COASTLINE)
cb = plt.colorbar(p, label = 'years', shrink = 0.8)
plt.title('Map of the mean time of carbon sequestration')
```

```python
# This is the map projection we want to plot *onto*
map_proj = ccrs.SouthPolarStereo()

p = fseq_OCIM2_48L_interpolated_SO['mean_seq_time'].mean(dim = 'depth').plot(
    transform=ccrs.PlateCarree(),  # the data's projection
    add_colorbar = False,
    subplot_kws={"projection": map_proj},
    #cmap="PuBu",
    #title = 'mean time of carbon sequesration'
)  # the plot's projection

#p.axes.coastlines()
p.axes.gridlines()
p.axes.add_feature(cfeature.LAND, color='grey')
p.axes.add_feature(cfeature.COASTLINE)
cb = plt.colorbar(p, label = 'years', shrink = 0.8)
plt.title('Map of the mean time of carbon sequestration')
```
